# Finding the top left corner

```python
import cv2
import numpy as np

original = cv2.imread("deskewed/1.jpg", cv2.IMREAD_GRAYSCALE)
_, binary = cv2.threshold(original, 127, 1, cv2.THRESH_BINARY_INV)

# slice up the images so I don't have to display them completely
x = 0
y = 0
w = 100
h = 100

binary_slice = binary[y:y+h, x:x+w]
original_slice = original[y:y+h, x:x+w]

cv2.imwrite("topleftcorner/binary_slice.jpg", binary_slice)
cv2.imwrite("topleftcorner/original_slice.jpg", original_slice)
```

```python
# examine the raw data to make sure it is what I want
for y in binary_slice:
    print(y)
```

```python
# find first y position
y_index = 0
for y in binary_slice:
    if any(y):
        print(y_index)
        break
    y_index += 1
```

```python
# find first x position
x_index = 0
for x in binary_slice[y_index]:
    if x == 1:
        print(x_index)
        break
    x_index += 1
```

```python
# double check the value 
binary_slice[y_index][x_index]
```

```python
# mark the corner so I know it's correct.
center = (x_index, y_index)
input_img = original_slice
small_radius = 2
big_radius = 6
color = (0,0,0) #black

circled = cv2.circle(input_img, center, small_radius, color)
circled = cv2.circle(input_img, center, big_radius, color)
cv2.imwrite("topleftcorner/circled.jpg", circled)
```
